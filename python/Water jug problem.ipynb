# BFS

```python
 graph = {
   '(0,0)': ['(5,0)','(0,3)'],
    '(5,0)' : ['(5,3)','(2,3)'],
    '(0,3)' : ['(3,0)'],
    '(5,3)' : [],
    '(2,3)' : ['(2,0)'],
    '(3,0)' : ['(3,3)'],
    '(2,0)' : ['(0,2)'],
     '(3,3)' : ['(5,1)'],
    '(0,2)' : ['(5,2)'],
     '(5,1)' : ['(0,1)'],
    '(5,2)' : ['(4,3)'],
     '(0,1)' : ['(1,1)'],
    '(4,3)' : ['(4,0)'],
    '(1,1)' : ['(1,3)'],
     '(4,0)' : [],
    '(1,3)' : ['(4,0)'],
     '(4,0)' : []
}
print(graph)
print(graph.keys())
print(graph.values())
print(graph.items())
for i,j in graph.items():
    print(i, ':' ,j)
visited = []
queue = []
def bfs(visited, graph, node):
    visited.append(node)
    queue.append(node)
    while queue:
        m = queue.pop(0)
        print(m,end = '')
        for neighbor in graph[m]:
            if neighbor not in visited:
                visited.append(neighbor)
                queue.append(neighbor)
print('BFS')
bfs(visited,graph,'(0,0)')
```

# DFS

```python
 graph = {
   '(0,0)': ['(5,0)','(0,3)'],
    '(5,0)' : ['(5,3)','(2,3)'],
    '(0,3)' : ['(3,0)'],
    '(5,3)' : [],
    '(2,3)' : ['(2,0)'],
    '(3,0)' : ['(3,3)'],
    '(2,0)' : ['(0,2)'],
     '(3,3)' : ['(5,1)'],
    '(0,2)' : ['(5,2)'],
     '(5,1)' : ['(0,1)'],
    '(5,2)' : ['(4,3)'],
     '(0,1)' : ['(1,1)'],
    '(4,3)' : ['(4,0)'],
    '(1,1)' : ['(1,3)'],
     '(4,0)' : [],
    '(1,3)' : ['(4,0)'],
     '(4,0)' : []
}
print(graph)
visited = set()
def dfs(visited,graph,node):
    if node not in visited:
        print(node)
        visited.add(node)
        for neighbor in graph[node]:
            dfs(visited, graph,neighbor)
print('DFS')
dfs(visited,graph,'(0,0)')
```
